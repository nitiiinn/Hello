```python
import numpy as npp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.DESCR
```

```python
iris.target 
```

```python
iris.target_names #output feature names
```

```python
name=iris.feature_names #feature column names 
```

```python
x=pd.DataFrame(iris['data'],columns=name)
```

```python
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
# plotting tthe decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
plt.show()
```

## post pruning


```python
model2=DecisionTreeClassifier(max_depth=2)
```

```python
model2.fit(x_train,y_train)
```

```python
# plotting tthe decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model2,filled=True)
plt.show()
```

```python
y_pred=model2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## pre pruning and hyperparameter tuning

```python
param_grid={'criterion':['gini','entropy','log_loss'],
        'splitter':['best','random'],
        'max_depth':[2,3,4,5,6,7,8,9,10],
        'max_features':['auto','sqrt','log2'],
        'min_samples_split':[2,3,4,5],
        'min_samples_leaf':[1,2,3,4]}
```

```python
model3=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(model3,param_grid,scoring='accuracy',cv=2,verbose=1,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred2=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```
