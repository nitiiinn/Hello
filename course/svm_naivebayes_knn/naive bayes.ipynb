```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
x,y=load_iris(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gauss=GaussianNB()
gauss.fit(x_train,y_train)
```

```python
y_pred=gauss.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print( accuracy_score(y_test, y_pred))
print( confusion_matrix(y_test, y_pred))
print( classification_report(y_test, y_pred))
```

## Tips dataset

```python
import seaborn as sns
df= sns.load_dataset('tips')
df.head(3)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
labelenc=LabelEncoder()
df['sex']=labelenc.fit_transform(df['sex'])
df['smoker']=labelenc.fit_transform(df['smoker'])
df['day']=labelenc.fit_transform(df['day'])
df['time']=labelenc.fit_transform(df['time'])
```

```python
df.head(3)
```

```python
df['day'].value_counts()
```

```python
x=df.drop(['smoker'],axis=1)
y=df['smoker']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
from sklearn.naive_bayes import GaussianNB
gauss=GaussianNB()
```

```python
# tried scaling but it isnt changing the accuracy in this case

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
gauss.fit(x_train_scaled, y_train)
```

```python
y_pred_tips=gauss.predict(x_test_scaled)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print( accuracy_score(y_test, y_pred_tips))
print( confusion_matrix(y_test, y_pred_tips))
print( classification_report(y_test, y_pred_tips))
```
