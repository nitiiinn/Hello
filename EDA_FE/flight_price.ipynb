```python
df.drop(['Route'], axis=1, inplace=True)
```

```python
df
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df
```

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
df
```

```python
df['Departure_hr']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]   
```

```python
df
```

```python
df.drop(['Dep_Time'], axis=1, inplace=True)
df
```

```python
df['Duration_hr']=df['Duration'].str.split('h').str[0]
df
```

```python
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df
```

```python
df.drop(['Duration'], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 1})
df
```

```python
df.info()
```

```python
df['Duration_hr'] = df['Duration_hr'].replace('5m', 0)
df['Duration_hr'] = df['Duration_hr'].astype(int)
```

```python
df.tail()
```

```python

df['Duration_min']=df['Duration_min'].replace('Nan', 0)
```

```python
df
```

```python

df['Duration_min']=df['Duration_min'].replace('Nan', 0 , inplace=True)


```

```python
df['Duration_hr'] = df['Duration_hr'].astype(int)
df['Duration_min'] = df['Duration_min'].fillna(0).astype(int)
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df['Departure_hr'] = df['Departure_hr'].astype(int)
df['Departure_min'] = df['Departure_min'].astype(int)
df.info()
```

```python
# Clean arrual_min column to keep only the minute part before any space
df['arrual_min'] = df['arrual_min'].str.split(' ').str[0]
```

```python
df
```

```python
# rename arrival columns
df.rename(columns={'arrual_min': 'Arrival_min', 'arrual_hr': 'Arrival_hr'}, inplace=True)
```

```python
df.info()
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray() 
encoded_df= pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Airline', 'Source', 'Destination']))
encoded_df.head()
df = pd.concat([df, encoded_df], axis=1)
```
